```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Insight: As the total bill increases, the tip amount also tends to increase.
# This indicates a positive relationship between bill amount and tipping behavior.

sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Relationship Between Total Bill and Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip Amount")
plt.show()
```

```python
# Insight: This graph shows the average tip given on different days.
# Usually, higher tips are observed during weekends (Saturday and Sunday).

sns.barplot(x="day", y="tip", data=tips)
plt.title("Average Tip Given on Different Days")
plt.xlabel("Day")
plt.ylabel("Average Tip")
plt.show()
```

```python
# Insight: This histogram shows the distribution of total bills.
# Most customers have bills between $10 and $20, indicating mid-range spending.

sns.histplot(tips["total_bill"], bins=10)
plt.title("Distribution of Total Bills")
plt.xlabel("Total Bill Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
# Shows number of rows and columns in the dataset
tips.shape
```

```python
# Check if the dataset has missing values
tips.isnull().sum()
```

```python
# Insight:
# This countplot shows the number of male and female customers in the dataset.

sns.countplot(x="sex", data=tips, palette="coolwarm")
plt.title("Customer Count by Gender")
plt.show()
```

```python
# Insight:
# Shows how total bills are distributed among customers.

sns.histplot(tips["total_bill"], bins=15, color="green")
plt.title("Distribution of Total Bills")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# Insight:
# Most tips are small, usually between $2 and $4.

sns.histplot(tips["tip"], bins=10, color="orange")
plt.title("Distribution of Tips")
plt.xlabel("Tip Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
# Insight:
# This graph compares tipping behavior between male and female customers.

sns.boxplot(x="sex", y="tip", data=tips, palette="pastel")
plt.title("Tip Amount by Gender")
plt.show()
```

```python
# Insight:
# This shows how the total bill varies depending on the day of the week.

sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title("Total Bill by Day")
plt.show()
```

```python
# Insight:
# This graph compares tips given by smokers and non-smokers.

sns.barplot(x="smoker", y="tip", data=tips, palette="viridis")
plt.title("Average Tip: Smokers vs Non-Smokers")
plt.show()
```

```python
# Insight:
# Larger groups tend to leave larger tips.

sns.scatterplot(x="size", y="tip", data=tips, color="purple")
plt.title("Party Size vs Tip")
plt.xlabel("Number of People")
plt.ylabel("Tip")
plt.show()
```

```python
# Insight:
# Heatmap shows correlation between numerical variables.

correlation = tips.corr(numeric_only=True)

sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
